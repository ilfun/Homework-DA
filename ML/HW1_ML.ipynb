```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
housing_df = pd.read_csv('housing.csv')
housing_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(housing_df.drop('MEDV',axis=1), housing_df['MEDV'], test_size=0.25, random_state=78)
```

```python
reg = LinearRegression().fit(X_train, y_train)
pred = reg.predict(X_test)

```

```python
mean_squared_error(y_test, pred)
```

```python
plt.figure(figsize=(20,9))
plt.plot(y_test.values, "blue",linewidth=5, label='Реальные значения')
plt.plot(pred, "red",linewidth=5, label='Предсказание')
plt.legend(loc='best')
plt.title("График пример")
plt.grid(True)
plt.show()
```
