```python
#очистка данных. название вакансий переведено в верхний регистр
vacs_copy2 =[{**row,'vactitle': row['vactitle'].upper()} for row in vacs_copy ]
```

```python
#очистка данных. описание вакансий переведено в верхний регистр
vacs_copy3 =[{**row,'vacdescription': row['vacdescription'].upper()} for row in vacs_copy2 ]
```

```python
#Поиск вакансий DA и вывод их количества
myDAwork = [row for row in vacs_copy3 if 'DATA ANALYST' in row['vactitle']]
len(myDAwork)
```

```python
myDAwork2 = [row for row in vacs_copy3 if 'АНАЛИТИК ДАННЫХ' in row['vactitle']]
len(myDAwork2)
```

```python
myDAwork3 = myDAwork + myDAwork2
len(myDAwork3)
#Всего вакансий на аналитика данных = вакансии которые мне нравятся
```

```python
myDAwork4 =[row for row in myDAwork3 if 'PYTHON' in row ['vacdescription']]
len(myDAwork4)
#Вакансий DA в которых требуется Python
```

```python
# Вакансии с моей текущей работой
myCURRENTwork = [row for row in vacs_copy3 if 'АНДЕРРАЙТЕР' in row['vactitle']]
len(myCURRENTwork)
```
